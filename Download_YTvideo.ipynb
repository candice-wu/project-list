```python
# "/Users/candicewu/git_project"
# https://www.youtube.com/watch?v=Ef-vc0c_-_g

def rbVideo():                                      # 使用者選取選項時的觸發函式 rbVideo
    global getvideo                                 # 預設值＝全域環境所設的變數值
    labelMsg.config(text = "")                      # 清除提示訊息的文字內容
    getvideo = videorb.get()                        # 取得使用者選取的選項按鈕的值

def clickDown():                                    # 使用者按下按鈕時的觸發函式 clickDown
    global getvideo                                 # 預設值＝全域環境所設的變數值
    labelMsg.config(text = "")                      # 清除提示訊息的文字內容
    if(url.get() == ""):                            # 判斷影片網址欄位值是否為空值
        labelMsg.config(text = "Please input URL")  # 利用config()方法去變更標籤元件的參數值
        return
    
    if(path.get() == ""):                           # 判斷存檔路徑是否為空值
        pathdir = "/Users/candicewu/Downloads"      # 預設存檔路徑
    else:
        pathdir = path.get()                        # 儲存至使用者輸入的存檔路徑中
    
    # 執行時，當使用者輸入的影片網址找不到符合條件的影片則會產生錯誤而中斷程式執行，
    # 故利用「try...except」做例外處理，而非中斷程式執行
    try:
        yt = YouTube(url.get())                     # 取得使用者輸入的影片網址
        
        # 依篩選條件下載影片，指定影片類型(subtype)＝mp4、解析度＝使用者選取的選項按鈕的值、影片格式須兼具影像(vcodec)及聲音(acodec)
        # 若有多個符合選項，則取第一個符合者並下載
        yt.streams.filter(subtype = "mp4", res = getvideo, progressive = True).first().download(pathdir)
        
        
        labelMsg.config(text = "影片下載完成✅" + time.ctime())      # 顯示「影片下載完成」訊息與下載時間
    
    except:
        
        labelMsg.config(text = "影片無法下載！" + time.ctime())      # 顯示「影片無法下載」訊息與失敗時間
    


import tkinter as tk                                # 自Python內建模組導入tkinter模組且命名為tk
from pytube import YouTube                          # 自pytube導入YouTube
import os                                           # 自Python內建模組導入os模組
import time                                         # 自Python內建模組導入time模組

win = tk.Tk()                                       # 建立主視窗名稱為 win
win.geometry("800x400")                             # 設定主視窗尺寸
win.title("Download YouTube Vedio")                 # 設定主視窗標題，若未設定則系統預設＝tk

url = tk.StringVar()                                # 建立文字型態的文字編輯元件的文字變數，供使用者輸入影片網址
path = tk.StringVar()                               # 建立文字型態的文字編輯元件的文字變數，供使用者輸入下載檔案的存檔路徑
videorb = tk.StringVar()                            # 建立文字型態的選項按鈕元件的文字變數，以儲存選項按鈕的值
getvideo = "360p"                                   # 預設儲存影片格式＝360p


# 建立上方的視窗區塊
frame_top = tk.Frame(win,                           # 指定主視窗標題
                     width = 600,                   # 設定視窗區塊的寬度
                     height = 120,                  # 設定視窗區塊的高度
                     background = "gray")           # 設定視窗區塊的背景顏色
frame_top.pack()                                    # 設定pack排版方式，將視窗區塊視為矩形物件顯示於指定的主視窗中

# 建立上方的視窗區塊 - 輸入網址的標籤元件
lbl_url = tk.Label(win,                             # 指定主視窗標題
                   text = "YouTube URL:")           # 設定標籤元件的文字內容
lbl_url.place(x = 120, y = 30)                      # 設定標籤元件的 x、y 座標

# 建立上方的視窗區塊 - 輸入網址的文字編輯元件
entry_url = tk.Entry(win,                           # 指定主視窗標題
                     width = 45,                    # 設定文字編輯元件的寬度
                     textvariable = url)            # 呼叫文字型態的文字編輯元件的文字變數 url
entry_url.place(x = 220, y = 30)                    # 設定文字編輯元件的 x、y 座標

# 建立上方的視窗區塊 - 輸入下載檔案的存檔路徑的標籤元件
lbl_path = tk.Label(win,                            # 指定主視窗標題
                    text = "Save Path:")            # 設定標籤元件的文字內容，預設值＝Python執行程式的資料夾
lbl_path.place(x = 120, y = 70)                     # 設定標籤元件的 x、y 座標

# 建立上方的視窗區塊 - 輸入下載檔案的存檔路徑的文字編輯元件
entry_path = tk.Entry(win,                          # 指定主視窗標題
                      width = 45,                   # 設定文字編輯元件的寬度
                      textvariable = path)          # 呼叫文字型態的文字編輯元件的文字變數 path
entry_path.place(x = 220, y = 70)                   # 設定文字編輯元件的 x、y 座標


# 建立下方的視窗區塊
frame_bottom = tk.Frame(win)                        # 指定主視窗標題
frame_top.pack()                                    # 設定pack排版方式，將視窗區塊視為矩形物件顯示於指定的主視窗中

# 建立下方的視窗區塊 - 下載影片的按鈕元件
btn_download = tk.Button(win,                       # 指定主視窗標題
                         text = "Download",         # 設定按鈕元件的文字內容
                         command = clickDown)       # 呼叫使用者按下按鈕時的觸發函式 clickDown
btn_download.place(x = 200, y = 130)                # 設定按鈕元件的 x、y 座標

# 建立下方的視窗區塊 - 影片格式的選項按鈕元件
rb1 = tk.Radiobutton(win,                           # 指定主視窗標題
                     text = "360p, mp4",            # 設定選項按鈕元件的文字內容
                     value = "360p",                # 設定使用者點選後的元件值
                     variable = videorb,            # 呼叫文字型態的選項按鈕元件的文字變數 videorb
                     command = rbVideo)             # 呼叫使用者選取選項時的觸發函式 rbVideo
rb1.place(x = 200, y = 170)                         # 設定選項按鈕元件的 x、y 座標
rb1.select()                                        # 預設選項按鈕元件選取rb1選項
rb2 = tk.Radiobutton(win,                           # 指定主視窗標題
                     text = "720p, mp4",            # 設定選項按鈕元件的文字內容
                     value = "720p",                # 設定使用者點選後的元件值
                     variable = videorb,            # 呼叫文字型態的選項按鈕元件的文字變數 videorb
                     command = rbVideo)             # 呼叫使用者選取選項時的觸發函式 rbVideo
rb2.place(x = 200, y = 200)                         # 設定選項按鈕元件的 x、y 座標


# 建立下方的視窗區塊 - 提示訊息的標籤元件
labelMsg = tk.Label(win,                            # 指定主視窗標題
                    text = "",                      # 設定標籤元件的文字內容，預設值＝空字串
                    foreground = "red")             # 設定標籤元件的文字顏色
labelMsg.place(x = 200, y = 240)                    # 設定標籤元件的 x、y 座標


win.mainloop()                                      # 讓程式進入與使用者互動模式，等待使用者觸發事件後進行處理
```
