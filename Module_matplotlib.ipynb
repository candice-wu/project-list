# 折線圖 plot

```python
import matplotlib.pyplot as plt                             # 導入matplotlib.pyplot模組，並取別名為plt

listx1 = [1, 5, 12, 19, 25, 36]
listy1 = [18, 50, 20, 38, 28, 42]
plt.plot(listx1, listy1, "g-*", ms=12, label="label 1")     # 繪出listx1與listy1串列，並設定線條與圖例

listx2 = [2, 6, 8, 11, 14, 16]
listy2 = [10, 40, 30, 50, 20, 40]
plt.plot(listx2, listy2, "r-s", ms=6, label="label 2")      # 繪出listx1與listy1串列，並設定線條與圖例

plt.title("Chart Title", fontsize=20)                       # 設定圖表標題與字體大小＝20
plt.xlabel("X-Label", fontsize=14)                          # 設定X座標名稱與字體大小＝14
plt.ylabel("Y-Label", fontsize=14)                          # 設定Y座標名稱與字體大小＝14

plt.xlim(0, 50)                                             # 設定X座標範圍
plt.ylim(0, 70)                                             # 設定Y座標範圍

plt.xticks(listx1)                                          # 設定X座標的刻度
#plt.tick_params(axis="both", labelsize=16, color="yellow") # 設定X、Y座標的刻度，字型＝16，顏色＝黃色

plt.grid(color="m", linestyle=":", linewidth=1, alpha=0.5)  # 設定格線顏色、線條樣式、線條寬度、透明度

plt.legend()
plt.show()
```

> **實例**
> - 繪製台北／台中自2015至2020年的銷售報表

```python
import matplotlib.pyplot as plt                                  # 導入matplotlib.pyplot模組，並取別名為plt

year = [2015, 2016, 2017, 2018, 2019, 2020]                      # 以 year年度 做為共用的 X座標資料串列
taipei = [128, 150, 199, 180, 150, 90]                           # Taipei 銷售數據
plt.plot(year, taipei, "b-.s", lw=2, ms=10, label="Taipei")      # 繪製 Taipei折線圖
taichung = [120, 145, 180, 170, 120, 105]                        # Taichung 銷售數據
plt.plot(year, taichung, "m--*", lw=2, ms=10, label="Taichung")  # 繪製 Taichung折線圖
plt.legend()                                                     # 顯示圖例 Taipei 與 Taichung

plt.ylim(0, 300)                                                 # 設定Y座標範圍 0~300，預設為100~200

plt.xticks(year)                                                 # 設定X座標以 year年度 做為刻度間隔

plt.title("Sales Report", fontsize=20)                           # 設定圖表標題＝Sales Report，且字體大小=20
plt.xlabel("Year", fontsize=12)                                  # 設定X座標名稱＝Year，且字體大小＝12
plt.ylabel("Million", fontsize=12)                               # 設定Y座標名稱＝Million，且字體大小＝12

plt.grid(color="black", linestyle=":", linewidth=1, alpha=0.5)   # 設定格線顏色＝黑色，線條樣式＝點狀線、線寬＝1，透明度＝0.5

plt.show()                                                       # 顯示圖表
```

# 長條圖 bar
> **類型一**

```python
import matplotlib.pyplot as plt

courses = ["C", "C++", "C#", "Java", "Python"]
people = [45, 28, 36, 78, 66]

# 設定每個矩形的寬度＝0.5，Y座標起始位置＝0，顏色rgb表示會依紅綠藍循環顯示
plt.bar(courses, people, width=0.5, bottom=0, color="rgb")

plt.title("程式語言修課人數表", fontsize=20)                     # 設定圖表標題
plt.xlabel("程式語言", fontsize=15)                            # 設定X座標名稱
plt.ylabel("修課人數", fontsize=15)                            # 設定Y座標名稱

#plt.rcParams["font.family"] = "Microsoft Sans Serif"         # 使用 Matplotlib模組中 rcParams函數，設定中文字型
#plt.rcParams["font.family"] = "Times New Roma"
plt.rcParams["font.family"] = "Arial"
#plt.rcParams["axes.unicode_minus"] = False

plt.show()
```

> **類型二：橫條圖 barh**

```python
import matplotlib.pyplot as plt

people = [45, 28, 36, 78, 66]                             # X座標：顯示每個項目的數值串列
courses = ["C", "C++", "C#", "Java", "Python"]            # Y座標：顯示每個項目的名稱串列或序列列


# 設定每個矩形的高度＝0.5，X座標起始位置＝0，顏色rgb表示會依紅綠藍循環顯示
plt.barh(courses, people, height=0.5, left=0, color="c")

plt.title("Enrollment Status Table", fontsize=20)         # 設定圖表標題
plt.xlabel("People", fontsize=15)                         # 設定X座標名稱
plt.ylabel("Programming Course", fontsize=15)                    # 設定Y座標名稱


plt.show()
```

> **類型三：堆疊長條圖 bar**

```python
import matplotlib.pyplot as plt

courses = ["C", "C++", "C#", "Java", "Python"]
man = [25, 20, 20, 18, 30]
female = [5, 18, 20, 15, 25]

# 設定每個矩形的寬度＝0.5，第二組長條圖的Y軸起始點設定為第一組的Y軸高度（終止值），並設定圖例
plt.bar(courses, man, width=0.5, label="Man")
plt.bar(courses, female, width=0.5, bottom=man, label="Female")
plt.legend()                                                          # 顯示圖例


plt.title("Enrollment Status Table", fontsize=20)                     # 設定圖表標題
plt.xlabel("Programming Course", fontsize=15)                         # 設定X座標名稱
plt.ylabel("People", fontsize=15)                                     # 設定Y座標名稱


plt.show()
```

# 圓餅圖 pie

```python
import matplotlib.pyplot as plt

size = [25, 30, 15, 10]
label = ["North", "West", "South", "East"]

color = ["blue", "green", "c", "m"]
explode = (0, 0, 0.2, 0)

plt.pie(size, 
        explode = explode, 
        labels = label, 
        colors = color, 
        labeldistance = 1.1,
        autopct = "%2.1f%%",
        pctdistance = 0.6,
        shadow = True,
        startangle = 90
       )

plt.show()
```

# 設定圖表區 figure

```python
import matplotlib.pyplot as plt

# 新增圖表區，不設定屬性採預設值
plt.figure()
plt.plot([1, 2, 3])


# 新增圖表區，設定屬性
plt.figure(figsize = [10, 5],
          dpi = 96,
          facecolor = "whitesmoke",
          edgecolor = "r",
          linewidth = 1,
          frameon = True,
          tight_layout = False
         )
plt.plot([1, 2, 3])

# 顯示圖表
plt.show()
```

# 圖表區中加入多張圖表 subplot
> **圖表區加入2列2欄的四張圖表**

```python
import matplotlib.pyplot as plt

plt.figure(figsize = [8, 8])

plt.subplot(2, 2, 1)
plt.title(label = "Chart 1", fontsize = 20)
plt.plot([1, 2, 3], "r:o")

plt.subplot(2, 2, 2)
plt.title(label = "Chart 2", fontsize = 20)
plt.plot([1, 2, 3], "g--o")

plt.subplot(2, 2, 3)
plt.title(label = "Chart 3", fontsize = 20)
plt.plot([1, 2, 3], "b:o")

plt.subplot(2, 2, 4)
plt.title(label = "Chart 1", fontsize = 20)
plt.plot([1, 2, 3], "y--o")

plt.show()
```

# 繪製台積電2020年度12月份最高/最低月成交價的月統計圖

```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

# 宣告月份、最高月成交價、最低月成交價的空字串
month, high, low = [], [], []

# 臺灣證券交易所-台積電的2020年度12月份月成交資訊的網址，指定給變數(url)
url = "https://www.twse.com.tw/exchangeReport/FMSRFK?response=html&date=20200101&stockNo=2330"

# 自訂HTTP Headers，可至 https://httpbin.org/get 查詢本機 User-Agent並複製貼上
headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/14.0.2 Safari/605.1.15",
}


# 將臺灣證券交易所-台積電的2020年度12月份月成交資訊指定給變數(content)
# 將自訂HTTP Headers加人Get請求中，並指派給新建立的response物件(content)
content = requests.get(url, headers = headers).text

# 利用lxml解析器去解析reguests模組取得的網頁原始碼，並指派給新建立的BeautifulSoup物件(sp)
sp = BeautifulSoup(content, "lxml")

# 取得第一個table
datas = sp.select("table")[0]

# 取得第一個div區塊內容並濾除空白文字
title = datas.find("div").text.replace(" ", "")

# 取得第一個tbody表格的tr表格列，共有12個
rows = datas.select("tbody tr")

# 逐一處理每列td表格
for row in rows:
    cols = row.select("td")
    month.append(int(cols[1].text))
    high.append(float(cols[2].text))
    low.append(float(cols[3].text))

# 繪製每月份的最高價／最低價的折線圖，與
plt.plot(month, high, "b-.s", lw = 2, ms = 5, label="Highest Price")
plt.plot(month, low, "m--*", lw = 2, ms = 5, label="Lowest Price")
plt.legend()

# 設定圖表標題、X座標名稱、Y座標名稱
plt.title(title, fontsize = 20)
plt.xlabel("Month", fontsize = 15)
plt.ylabel("Price", fontsize = 15)


# 使用 Matplotlib模組中 rcParams函數，設定中文字型
plt.rcParams["font.family"] = "Arial"
# plt.rcParams["font.sans-serif"] = "DFkai-SB"

# 設定格線顏色＝黑色，線條樣式＝點狀線、線寬＝1，透明度＝0.5
plt.grid(color="black", linestyle=":", linewidth=1, alpha=0.5)

# 顯示圖表
plt.show()
```
